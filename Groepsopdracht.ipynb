```python
airlines.columns = airlines.columns.str.replace('IATA_CODE', 'AIRLINE_CODE')
flights.columns = flights.columns.str.replace('AIRLINE', 'AIRLINE_CODE')
airlines.head()
```

```python
flights_airlines_merge = pd.merge(flights, airlines, how='outer', on='AIRLINE_CODE')
flights_airlines_merge.head()
```

```python
origin_airport = pd.read_csv('airports.csv')
destination_airport = pd.read_csv('airports.csv')
origin_airport.set_axis(["ORIGIN_AIRPORT", "ORIGIN_AIRPORT_NAME", "ORIGIN_AIRPORT_CITY", "ORIGIN_AIRPORT_STATE", "ORIGIN_AIRPORT_COUNTRY", "ORIGIN_AIRPORT_LATITUDE", "ORIGIN_AIRPORT_LONGITUDE"], axis="columns", inplace=True)
destination_airport.set_axis(['DESTINATION_AIRPORT', 'DESTINATION_AIRPORT_NAME', 'DESTINATION_AIRPORT_CITY', 'DESTINATION_AIRPORT_STATE', 'DESTINATION_AIRPORT_COUNTRY', 'DESTINATION_AIRPORT_LATITUDE', 'DESTINATION_AIRPORT_LONGITUDE'], axis='columns', inplace=True)
```

```python
origin_airport.head()
```

```python
destination_airport.head()
```

```python
airport_origin_merge = pd.merge(flights_airlines_merge, origin_airport, how='outer', on='ORIGIN_AIRPORT')
airport_destination_merge = pd.merge(airport_origin_merge, destination_airport, how='outer', on='DESTINATION_AIRPORT')
airport_destination_merge.head()
```

```python
airport_destination_merge.nunique()
```

```python
nan_count = airport_destination_merge.isna().sum()
nan_count
```

```python
airport_destination_merge.tail()
```

```python
columns_drop = ['ORIGIN_AIRPORT_COUNTRY', 'ORIGIN_AIRPORT_LATITUDE','ORIGIN_AIRPORT_LONGITUDE', 'DESTINATION_AIRPORT_COUNTRY', 'DESTINATION_AIRPORT_LATITUDE', 'DESTINATION_AIRPORT_LONGITUDE', 'TAXI_OUT', 'WHEELS_OFF', 'DISTANCE', 'WHEELS_ON', 'TAXI_IN', 'AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_CODE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']
airport_drop = airport_destination_merge.drop(columns=columns_drop, axis = 1)
airport_drop.head(20)
```

```python
airport_drop.info()
```

```python
dag_mapping = {1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday', 7:'Sunday'}
airport_drop['DAY_OF_WEEK'] = airport_drop['DAY_OF_WEEK'].replace(dag_mapping)
airport_drop.head()
```

```python
airport_drop['DATE'] = airport_drop[['MONTH', 'DAY', 'YEAR']].astype(str).apply(lambda row: '-'.join(row), axis=1)
airport_drop.head()
```

```python
kolom = airport_drop.pop('DATE')
airport_drop.insert(0, 'DATE', kolom)
airport_drop.head()
```

```python
date_drop = ['YEAR', 'MONTH', 'DAY']
airport_new = airport_drop.drop(columns=date_drop, axis = 1)
airport_new.head()
```

```python
airport_new.info()
```

```python
nan_count = airport_new.isna().sum()
nan_count
```

```python
#airport_new = airport_new.fillna(0)

airport_new['DEPARTURE_TIME'] = airport_new['DEPARTURE_TIME'].astype(int)
airport_new['DEPARTURE_DELAY'] = airport_new['DEPARTURE_DELAY'].astype(int)
airport_new['SCHEDULED_TIME'] = airport_new['SCHEDULED_TIME'].astype(int)
airport_new['ELAPSED_TIME'] = airport_new['ELAPSED_TIME'].astype(int)
airport_new['AIR_TIME'] = airport_new['AIR_TIME'].astype(int)
airport_new['ARRIVAL_TIME'] = airport_new['ARRIVAL_TIME'].astype(int)
```

```python
airport_new['TAIL_NUMBER'] = airport_new['TAIL_NUMBER'].astype(str)
airport_new['ORIGIN_AIRPORT'] = airport_new['ORIGIN_AIRPORT'].astype(str)
airport_new['DESTINATION_AIRPORT'] = airport_new['DESTINATION_AIRPORT'].astype(str)
airport_new['AIRLINE'] = airport_new['AIRLINE'].astype(str)
airport_new['ORIGIN_AIRPORT_NAME'] = airport_new['ORIGIN_AIRPORT_NAME'].astype(str)
airport_new['ORIGIN_AIRPORT_CITY'] = airport_new['ORIGIN_AIRPORT_CITY'].astype(str)
airport_new['ORIGIN_AIRPORT_STATE'] = airport_new['ORIGIN_AIRPORT_STATE'].astype(str)
airport_new['DESTINATION_AIRPORT_NAME'] = airport_new['DESTINATION_AIRPORT_NAME'].astype(str)
airport_new['DESTINATION_AIRPORT_CITY'] = airport_new['DESTINATION_AIRPORT_CITY'].astype(str)
airport_new['DESTINATION_AIRPORT_STATE'] = airport_new['DESTINATION_AIRPORT_STATE'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
for kolom in airport_new.columns:     
    if airport_new[kolom].dtype == 'object':         
        airport_new[kolom] = le.fit_transform(airport_new[kolom])
```

```python
from sklearn.preprocessing import LabelEncoder

df = airport_new.copy()
le = LabelEncoder()
df['DATE'] = le.fit_transform(df['DATE'])
df['DAY_OF_WEEK'] = le.fit_transform(df['DAY_OF_WEEK'])
df['AIRLINE_CODE'] = le.fit_transform(df['AIRLINE_CODE'])
df['FLIGHT_NUMBER'] = le.fit_transform(df['FLIGHT_NUMBER'])
df['TAIL_NUMBER'] = le.fit_transform(df['TAIL_NUMBER'])
df['ORIGIN_AIRPORT'] = le.fit_transform(df['ORIGIN_AIRPORT'])
df['DESTINATION_AIRPORT'] = le.fit_transform(df['DESTINATION_AIRPORT'])
df['SCHEDULED_DEPARTURE'] = le.fit_transform(df['SCHEDULED_DEPARTURE'])
df['DEPARTURE_TIME'] = le.fit_transform(df['DEPARTURE_TIME'])
df['DEPARTURE_DELAY'] = le.fit_transform(df['DEPARTURE_DELAY'])
df['SCHEDULED_TIME'] = le.fit_transform(df['SCHEDULED_TIME'])
df['ELAPSED_TIME'] = le.fit_transform(df['ELAPSED_TIME'])
df['AIR_TIME'] = le.fit_transform(df['AIR_TIME'])
df['SCHEDULED_ARRIVAL'] = le.fit_transform(df['SCHEDULED_ARRIVAL'])
df['ARRIVAL_TIME'] = le.fit_transform(df['ARRIVAL_TIME'])
df['AIRLINE'] = le.fit_transform(df['AIRLINE'])
df['ORIGIN_AIRPORT_NAME'] = le.fit_transform(df['ORIGIN_AIRPORT_NAME'])
df['ORIGIN_AIRPORT_CITY'] = le.fit_transform(df['ORIGIN_AIRPORT_CITY'])
df['ORIGIN_AIRPORT_STATE'] = le.fit_transform(df['ORIGIN_AIRPORT_STATE'])
df['DESTINATION_AIRPORT_NAME'] = le.fit_transform(df['DESTINATION_AIRPORT_NAME'])
df['DESTINATION_AIRPORT_CITY'] = le.fit_transform(df['DESTINATION_AIRPORT_CITY'])
df['DESTINATION_AIRPORT_STATE'] = le.fit_transform(df['DESTINATION_AIRPORT_STATE'])
```

```python
from sklearn.model_selection import train_test_split

y = df['DEPARTURE_DELAY']
X = df[['ORIGIN_AIRPORT', 'AIRLINE']]

X_train, X_test, y_train, y_test = train_test_split(
                                                X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
# predict the scores based on the test data
y_pred = clf.predict(X_test)

#compute the accuracy score by comparing the y_test results with the predicted results.
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
