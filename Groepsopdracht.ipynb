```python
airlines.columns = airlines.columns.str.replace('IATA_CODE', 'AIRLINE_CODE')
flights.columns = flights.columns.str.replace('AIRLINE', 'AIRLINE_CODE')
airlines.head()
```

```python
flights_airlines_merge = pd.merge(flights, airlines, how='outer', on='AIRLINE_CODE')
flights_airlines_merge.head()
```

```python
origin_airport = pd.read_csv('airports.csv')
destination_airport = pd.read_csv('airports.csv')
origin_airport.set_axis(["ORIGIN_AIRPORT", "ORIGIN_AIRPORT_NAME", "ORIGIN_AIRPORT_CITY", "ORIGIN_AIRPORT_STATE", "ORIGIN_AIRPORT_COUNTRY", "ORIGIN_AIRPORT_LATITUDE", "ORIGIN_AIRPORT_LONGITUDE"], axis="columns", inplace=True)
destination_airport.set_axis(['DESTINATION_AIRPORT', 'DESTINATION_AIRPORT_NAME', 'DESTINATION_AIRPORT_CITY', 'DESTINATION_AIRPORT_STATE', 'DESTINATION_AIRPORT_COUNTRY', 'DESTINATION_AIRPORT_LATITUDE', 'DESTINATION_AIRPORT_LONGITUDE'], axis='columns', inplace=True)
```

```python
origin_airport.head()
```

```python
destination_airport.head()
```

```python
airport_origin_merge = pd.merge(flights_airlines_merge, origin_airport, how='outer', on='ORIGIN_AIRPORT')
airport_destination_merge = pd.merge(airport_origin_merge, destination_airport, how='outer', on='DESTINATION_AIRPORT')
airport_destination_merge.head()
```

```python
airport_destination_merge.nunique()
```
